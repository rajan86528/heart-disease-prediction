# matplotlib example workflow

```python
%matplotlib inline
import matplotlib.pyplot as plt

#1. prepare data
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]

# 2. setup plot
fig, ax = plt.subplots(figsize=(10, 10))

# 3. Plot data
ax.plot(x, y)

# 4. Customize plot
ax.set(title = "simple plot",
      xlabel = "x-axis",
      ylabel = "y-axis")
# 5.Save  & show
fig.savefig("sample.png")
```

```python
import numpy as np
```

```python
x = np.linspace(0, 10, 100)
x[:10]
```

```python
# Plot the data and create line plot
fig, ax = plt.subplots()
ax.plot(x, x**2)
```

```python
# Scatter plot

fig, ax = plt.subplots()
ax.scatter(x, np.exp(x))
```

```python
# Scatter plot

fig, ax = plt.subplots()
ax.scatter(x, np.sin(x))
```

```python
# Make plot using dict

fruit_price = {'mango':25, "orange":50, "apple":70, 'banana':94}

fig, ax = plt.subplots()
ax.bar(fruit_price.keys(),fruit_price.values())
ax.set(title="Price of Fruit", ylabel="Price of fruit", xlabel="Name of fruit")
```

```python
fig, ax = plt.subplots()
ax.barh(list(fruit_price.keys()), list(fruit_price.values()))
```

```python
x = np.random.randn(1000)
fig, ax = plt.subplots()
ax.barh(list(fruit_price.keys()), list(fruit_price.values()))
```

# Subplots Option 1

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows = 2, ncols=2, figsize=(10, 5))

ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(fruit_price.keys(), fruit_price.values())
ax4.hist(np.random.randn(1000))
```

# Subplots Option 2

```python
fix, ax = plt.subplots(nrows=2, ncols=2, figsize=(10, 5))

ax[0, 0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(fruit_price.keys(), fruit_price.values())
ax[1,1].hist(np.random.randn(1000))
```

```python
#  oo method
```

```python
over_50 = heart_disease[heart_disease['age']>50]
```

```python
over_50.head()
```

```python
over_50.plot(kind="scatter",x = 'age', y='chol', c= 'target')
```

```python
fix, ax = plt.subplots(figsize=(10, 6))
over_50.plot(kind="scatter",x = 'age', y='chol', c= 'target', ax=ax)
ax.set_xlim([45, 100])
```

```python
fix, ax = plt.subplots(figsize=(10, 6))
scatter  =ax.scatter(x=over_50['age'],
                    y=over_50['chol'],
                    c=over_50['target'])
ax.set(title='Heart Disease and cholestrol Levels',
      xlabel="Age",
      ylabel='Cholesterol')

ax.legend(*scatter.legend_elements(), title='Target')

ax.axhline(over_50["chol"].mean(),
          linestyle='--')
```

```python
fig, (ax0, ax1) = plt.subplots(nrows=2, ncols=1, figsize=(10, 10),
                               sharex=True)


scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])
ax0.set(title="Heart Disease and Cholestrol Levels",
#        xlabel="Age",
       ylabel="cholesterol")
ax0.legend(*scatter.legend_elements(), title='Target')
ax0.axhline(y=over_50['chol'].mean(),
           linestyle="--")


scatter = ax1.scatter(x=over_50['age'],
                     y=over_50['thalach'],
                     c=over_50['target'])
ax1.set(title="Heart Disease and max Heart Rate",
       xlabel="Age",
       ylabel="max Heart Rate")
ax1.legend(*scatter.legend_elements(), title='Target')
ax1.axhline(y=over_50['thalach'].mean(),
           linestyle="--")





fig.suptitle("Heart Disease Analysis", fontsize=16, fontweight='bold')
```
